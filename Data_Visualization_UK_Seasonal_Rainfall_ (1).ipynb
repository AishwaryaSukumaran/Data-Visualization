Step 1: Install & Import Required Libraries

```python
# Install required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

Step 2: Define File Path & Load Data

```python
# Define the file path (Update this to your Google Drive path)
file_path = "/content/drive/MyDrive/Data_visualization  /UK_rainfall.xlsx"

# Read the Excel file from Google Drive
df_UK = pd.read_excel(file_path)
```

Step 3: Data Processing

```python
# Filter data for years 2000-2022 and select required columns
df_seasonal = df_UK[(df_UK["year"] >= 2000) & (df_UK["year"] <= 2022)][["year", "win", "spr", "sum", "aut"]]

# Calculate max and min seasonal rainfall
df_seasonal["Maximum_rainfall"] = df_seasonal[["win", "spr", "sum", "aut"]].max(axis=1)
df_seasonal["Minimum_rainfall"] = df_seasonal[["win", "spr", "sum", "aut"]].min(axis=1)

# Rename columns for clarity
df_seasonal.rename(columns={"year": "Year", "win": "Winter", "spr": "Spring",
                            "sum": "Summer", "aut": "Autumn"}, inplace=True)

# Display processed data
df_seasonal.head()
```

Step 4: Plot Max & Min Seasonal Rainfall

```python
# Set figure size
plt.figure(figsize=(14, 6))

# Plot max and min seasonal rainfall
plt.plot(df_seasonal["Year"], df_seasonal["Maximum_rainfall"], marker="o", linestyle="-", linewidth=2, color="blue", label="Max Rainfall")
plt.plot(df_seasonal["Year"], df_seasonal["Minimum_rainfall"], marker="s", linestyle="--", linewidth=2, color="red", label="Min Rainfall")

# Customizing the plot
plt.title("Max and Min Seasonal Rainfall in UK (2000-2022)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12, fontweight='bold')
plt.ylabel("Rainfall (mm)", fontsize=12, fontweight='bold')
plt.xticks(df_seasonal["Year"], rotation=45)
plt.yticks(range(0, int(df_seasonal["Maximum_rainfall"].max()) + 20, 20))  # Y-axis starts from 0
plt.grid(True, linestyle="--", alpha=0.6)  # Add grid lines
plt.legend()

# Show the plot
plt.show()
```

Step 4: Plot Seasonal Rainfall Trends

```python
# Set figure size
plt.figure(figsize=(14, 6))

# Plot seasonal rainfall trends
plt.plot(df_seasonal["Year"], df_seasonal["Winter"], marker="o", linestyle="-", linewidth=2, color="cyan", label="Winter")
plt.plot(df_seasonal["Year"], df_seasonal["Spring"], marker="s", linestyle="--", linewidth=2, color="green", label="Spring")
plt.plot(df_seasonal["Year"], df_seasonal["Summer"], marker="^", linestyle=":", linewidth=2, color="orange", label="Summer")
plt.plot(df_seasonal["Year"], df_seasonal["Autumn"], marker="d", linestyle="-.", linewidth=2, color="brown", label="Autumn")

# Customizing the plot
plt.title("Seasonal Rainfall Trends in UK (2000-2022)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12, fontweight='bold')
plt.ylabel("Rainfall (mm)", fontsize=12, fontweight='bold')
plt.xticks(df_seasonal["Year"], rotation=45)
plt.yticks(range(0, int(df_seasonal[["Winter", "Spring", "Summer", "Autumn"]].max().max()) + 20, 20))  # Y-axis starts from 0
plt.grid(True, linestyle="--", alpha=0.6)  # Add grid lines
plt.legend()

# Show the plot
plt.show()
```
